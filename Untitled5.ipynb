```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Load the trained model
model = tf.keras.models.load_model('my_model.h5')

# Specify the class names (same as the ones used when creating the model)
class_names = ['Bacterialblight', 'Blast', 'Brownspot', "Healthy", "Tungro"]

# Function to preprocess the image for prediction (resize and normalize)
def preprocess_image(image_path, target_size=(64, 64)):
    # Load the image from the file path
    image = Image.open(image_path)
    image = image.convert('RGB')  # Convert to RGB if it's not
    image = image.resize(target_size)  # Resize image to match model input
    image = np.array(image) / 255.0  # Normalize image
    return np.expand_dims(image, axis=0)  # Add batch dimension

# Function to make prediction and display result
def predict_disease(image_path, model, class_names):
    # Preprocess the image for prediction
    processed_image = preprocess_image(image_path)
    
    # Make prediction
    prediction = model.predict(processed_image)
    predicted_class_idx = np.argmax(prediction, axis=1)
    predicted_class = class_names[predicted_class_idx[0]]
    confidence = prediction[0][predicted_class_idx[0]]
    
    # Display the image and prediction
    plt.figure(figsize=(8, 6))
    img = plt.imread(image_path)
    plt.imshow(img)
    plt.title(f"Predicted: {predicted_class}\nConfidence: {confidence:.2f}")
    plt.axis('off')
    plt.show()
    
    # Return the predicted class and confidence
    return predicted_class, confidence

# Example usage: Predict disease from a new image
image_path = 'temp.jpg'  # Replace with your image file path
predicted_class, confidence = predict_disease(image_path, model, class_names)
print(f"Predicted class: {predicted_class}")
print(f"Confidence: {confidence:.2f}")
```
